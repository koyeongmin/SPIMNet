```python
# step1: extract positive and negative pairs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2

path = '/home/yeongmin/shared_dir/research/SPIMNet/NIR_RGB/'
categories = ['country', 'field', 'indoor', 'oldbuilding', 'street', 'urban', 'water', 'forest', 'mountain']

for cate in categories:

    csv_path = path+'image-datasets/csv-files/'+cate+'.csv'
    data = pd.read_csv(csv_path, skiprows=0)

    type_data = data['type']

    rgb_fname = data['rgb']
    rgb_xs    = data['rgb_x']
    rgb_ys    = data['rgb_y']

    nir_fname = data['nir']
    nir_xs    = data['nir_x']
    nir_ys    = data['nir_y']

    length = type_data.count()
    print(cate+'-length: ', length)

    # for crop image
    side = 32
    pos_index = 0
    neg_index = 0

    for i in range(length):
        # get data for a row
        s_type = type_data.iloc[i]

        s_rgb_fname = rgb_fname.iloc[i]
        s_rgb_x = rgb_xs.iloc[i]
        s_rgb_y = rgb_ys.iloc[i]

        s_nir_fname = nir_fname.iloc[i]
        s_nir_x = nir_xs.iloc[i]
        s_nir_y = nir_ys.iloc[i]  

        # read images
        s_rgb_fname = path + 'image-datasets/'+cate+'/' + s_rgb_fname
        s_nir_fname = path + 'image-datasets/'+cate+'/' + s_nir_fname
        
        img_rgb = cv2.imread(s_rgb_fname, 1)
        img_nir = cv2.imread(s_nir_fname, 1)

        # crop
        patch_rgb = img_rgb[s_rgb_y-side:s_rgb_y+side, s_rgb_x-side:s_rgb_x+side]
        patch_nir = img_nir[s_nir_y-side:s_nir_y+side, s_nir_x-side:s_nir_x+side]    

        img_list  = [patch_rgb, patch_nir]
        imgs_comb = np.hstack(img_list)
        
        if s_type == 'negative':
            cv2.imwrite(path+'patch-datasets/'+cate+'/'+str(neg_index)+'_neg.jpg', imgs_comb)
            neg_index = neg_index + 1     
        else:
            cv2.imwrite(path+'patch-datasets/'+cate+'/'+str(pos_index)+'_pos.jpg', imgs_comb)
            pos_index = pos_index + 1

    print('neg_index: ', neg_index)
    print('pos_index: ', pos_index)
    print('')
```
